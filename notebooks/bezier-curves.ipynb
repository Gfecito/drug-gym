```python
def cubic_bezier(p0, p1, p2, p3, t):
    """Compute the position on a cubic Bezier curve by parameter t."""
    return ((1 - t)**3 * np.array(p0) + 
            3 * (1 - t)**2 * t * np.array(p1) +
            3 * (1 - t) * t**2 * np.array(p2) + 
            t**3 * np.array(p3))

# Bezier curve control points
points = np.array([[12, 4], [18, 3]])

def plot_edge(points):
    p0 = points[0]
    p1 = np.array([points[0][0], points[0][1] - 0.5])
    p2 = np.array([points[1][0], points[1][1] + 0.5])
    p3 = points[1]

    # Generate t values
    t_values = np.linspace(0, 1, 500)

    # Calculate the points on the Bezier curve
    bezier_points = np.array([cubic_bezier(p0, p1, p2, p3, t) for t in t_values])

    # Plotting the Bezier curve with control points
    plt.figure(figsize=(10, 6))
    plt.plot(bezier_points[:, 0], bezier_points[:, 1], label='Cubic Bezier Curve', color='grey')
    plt.title('Cubic Bezier Curve through Given Points')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
plot_edge(points)
```
