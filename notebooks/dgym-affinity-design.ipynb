```python
%reload_ext autoreload
%autoreload 2

import rdkit
rdkit.Chem.Draw.IPythonConsole.ipython_maxProperties = -1

import dgym as dg

# load all data
print('load data')
path = '../../dgym-data'

deck = dg.MoleculeCollection.load(
    f'{path}/DSi-Poised_Library_annotated.sdf',
    reactant_names=['reagsmi1', 'reagsmi2', 'reagsmi3']
)

reactions = dg.ReactionCollection.from_json(
    path = f'{path}/All_Rxns_rxn_library.json',
    smarts_col = 'reaction_string',
    classes_col = 'functional_groups'
)

building_blocks = dg.datasets.disk_loader(f'{path}/Enamine_Building_Blocks_Stock_262336cmpd_20230630.sdf')
fingerprints = dg.datasets.fingerprints(f'{path}/out/Enamine_Building_Blocks_Stock_262336cmpd_20230630.fpb')

# align fingerprints to building blocks
print('align fingerprints')
fingerprints = dg.utils.sort_fingerprints(fingerprints, building_blocks)
```

```python
import numpy as np
from dgym.envs.library_designer import LibraryDesigner
from dgym.envs.oracle import DGLOracle, RDKitOracle, DockingOracle
from dgym.envs.evaluator import ClassicEvaluator, UtilityFunction

# create library designer
library_designer = LibraryDesigner(
    reactions,
    building_blocks,
    fingerprints
)

# create docking oracle
config = {
    'center_x': 9.812,
    'center_y': -0.257,
    'center_z': 20.8485,
    'size_x': 14.328,
    'size_y': 8.85,
    'size_z': 12.539,
    'exhaustiveness': 128,
    'max_step': 20,
    'num_modes': 9,
    'scoring': 'vinardo',
    'refine_step': 3,
    'seed': 5
}

docking_oracle = DockingOracle(
    'Mpro affinity',
    receptor_path=f'../../dgym-data/Mpro_prepped.pdbqt',
    config=config
)
```

```python
library_designer.design(
    dg.molecule.Molecule('O=C(c1ccc2c(=O)[nH][nH]c(=O)c2c1)n1nnc2ccccc2c1=O',
                         reactants = []),
    size=10,
    temperature=0.0
)
```

```python
next(library_designer.generate_analogs([], temperature=0.0))[0]
```

```python
import numpy as np
from tqdm.notebook import tqdm

best_mol = dg.molecule.Molecule('O=C(c1ccc2c(=O)[nH][nH]c(=O)c2c1)n1nnc2ccccc2c1=O',
                    reactants=[dg.molecule.Molecule('O=C(O)c1ccc2c(=O)[nH][nH]c(=O)c2c1'),
                               dg.molecule.Molecule('O=c1[nH]nnc2ccccc12')])
best_score = docking_oracle([best_mol])

mols = dg.collection.MoleculeCollection([best_mol])
all_scores = [best_score]

for _ in tqdm(range(10)):
    
    display(best_mol)
    print(best_score)
    
    analogs = library_designer.design(best_mol, temperature=0.0, size=50)
    mols += analogs

    scores = docking_oracle(analogs)
    all_scores.append(scores)
    
    best = np.nanargmax(scores).item()
    best_score = scores[best]
    best_mol = analogs[best]
```
